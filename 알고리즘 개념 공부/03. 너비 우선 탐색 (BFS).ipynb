```python
# extend 
data = [1, 2, 3]
data.extend([4, 5])

data
```

```python
def bfs(graph, start_node):
    visited = list()
    need_visit = list()
    
    need_visit.append(start_node)
    
    while need_visit:
        print('need_visit : ', need_visit)
        print('visited : ', visited , '\n')
        node = need_visit.pop(0)
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])
            
    return visited
```

```python
bfs(graph, 'A')
```

    -   : O(V+E)

```python
def bfs(graph, start_node):
    visited = list()
    need_visit = list()
    
    need_visit.append(start_node)
    count = 0
    while need_visit:
        count += 1
        node = need_visit.pop(0)
        
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])
    print(count)

    return visited
```

```python
bfs(graph, 'A')
```
