```python
numbers = "17"

def solution(numbers):
    list(numbers)
    
    answer = []
    
    from itertools import permutations, chain
    
    comb_num_list = []
    
    for n in range(1, len(numbers)+2):
        l = list(permutations(numbers, n))
        l = [''.join(list(i)) for i in l]
        comb_num_list.append(l)
    
    comb_num_list = list(chain.from_iterable(comb_num_list))
    comb_num_list = list(set([int(i) for i in comb_num_list]))
    
    print(comb_num_list)
    
    #    1  
    answer = len(comb_num_list)
    
    for num in comb_num_list:
        if num == 1:
            answer -= 1
            continue
        
        for n in range(2, int(num**0.5) + 1):
            
            if (num % n == 0) : # 
                answer -= 1
    
    
    return answer
    
solution(numbers)
```

```python
import math
import itertools

#   
def is_decimal(n):
    if n < 2: return False

    to = int(math.sqrt(n)) + 1
    for i in range(2, to):
        if n % i == 0: 
            return False
    return True

def solution(number):
    candidate = set()

    for i in range(len(number)):
        numbers = set(map(int, map(''.join, itertools.permutations(number, i+1))))
        candidate |= numbers # 

    answer = 0
    candidate = list(candidate) #  
    for n in candidate:
        if is_decimal(n):
            answer += 1
    return answer
```
