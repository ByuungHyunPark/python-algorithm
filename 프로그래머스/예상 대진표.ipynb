```python
def solution(n,a,b):
    import math
    
    answer = 0

    r = 0
    
    while True:
        a_turn , b_turn = int(math.ceil(a % (n+1))) , int(math.ceil(b % (n+1)))
        
#        print('\na_turn : ', a_turn)
#        print('b_turn : ', b_turn)
        
        a = a/2
        b = b/2
        
        n = int(n/2)
    
        if a_turn == b_turn:
            return r
        
        #다음 단계로
        r += 1
        
solution(15, 1, 15)
```
