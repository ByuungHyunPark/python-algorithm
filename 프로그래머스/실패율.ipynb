```python
N = 6
stages = [2, 1, 2, 6, 2, 4, 3, 3]

def solution(N, stages):
    import collections
    
    answer = []
    
    
    stages = sorted(stages)
    stage_dict = dict()
 
    for i in range(1, N+1):
        len_stages = len(stages)
        fail_list = [k for k in stages if k == i] #stages list i 
        
        if len_stages == 0 :
            stage_dict[i] = 0
        else:
            stage_dict[i] =  len(fail_list) / len_stages
        
        stages = [k for k in stages if k != i] #drop
        
    
    return [j[0]for j in sorted(stage_dict.items(), key = lambda x : x[1], reverse = True)]

solution(N, stages)
```
