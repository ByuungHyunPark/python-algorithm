생각보다 쉽게 풀이된듯.
- 런타임오류가 발생해서 원인을 찾아보니, len_stages == 0 일때, 실패율을 계산하는 과정에서 division_zero 오류로 인해 발생 => len_stages == 0 조건문 따로 생성 후 해결

```python
N = 6
stages = [2, 1, 2, 6, 2, 4, 3, 3]

def solution(N, stages):
    import collections
    
    answer = []
    
    
    stages = sorted(stages)
    stage_dict = dict()
 
    for i in range(1, N+1):
        len_stages = len(stages)
        fail_list = [k for k in stages if k == i] #stages list중 i가 아닌애들만
        
        if len_stages == 0 :
            stage_dict[i] = 0
        else:
            stage_dict[i] =  len(fail_list) / len_stages
        
        stages = [k for k in stages if k != i] #drop
        
    
    return [j[0]for j in sorted(stage_dict.items(), key = lambda x : x[1], reverse = True)]

solution(N, stages)
```
