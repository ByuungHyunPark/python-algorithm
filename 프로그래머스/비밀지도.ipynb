```python
n = 5
arr1 = [9, 20, 28, 18, 11]
arr2 = [30, 1, 21, 17, 28]
#	["#####","# # #", "### #", "# ##", "#####"]

def solution(n, arr1, arr2):
    
    answer = []

    # 2  
    arr1 = [str(bin(i))[2:] for i in arr1]
    for index, i in enumerate(arr1):
        arr1[index] = '0' * (n - len(i)) + i
    
    arr2 = [str(bin(i))[2:] for i in arr2]
    for index, i in enumerate(arr2):
        arr2[index] = '0' * (n - len(i)) + i    
    
    #######################################
    for i,j in zip(arr1, arr2):
        list_ = []
        for k, l in zip(list(i), list(j)):
            
            
            if int(k) | int(l) == 0:
                list_.append(' ')
            else:
                list_.append('#')
        answer.append(''.join(list_))
    
    
    return answer


solution(n, arr1, arr2)
```
