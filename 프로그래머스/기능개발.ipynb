```python
def solution(progresses, speeds):
    answer = []
    
    import math
    import numpy as np
    
    need_day_list = [math.ceil((100 - pr) /s) for pr, s in zip(progresses, speeds)]
    now_max = 0
    cnt = 1
    
    max_idx = []
    for idx, i in enumerate(need_day_list):
        if i > now_max:
            now_max = i
            max_idx.append(idx)
    
    l1 = max_idx
    l2 = max_idx[1:] + [len(progresses)]
    
    return list(np.array(l2) - np.array(l1))

solution([95, 90, 99, 99, 80, 99], [1, 1, 1, 1, 1, 1])
```

```python
def solution(progresses, speeds):
    import math
    answer=[]

    remaining = [100 - p for p in progresses]
    print("remaining: ", remaining)
    remaining_days = [math.ceil(r / s) for r, s in zip(remaining,speeds)]
    print("remaining_days: ", remaining_days)


    cnt = 0
    i = 0

    while i < len(remaining_days):
        cnt += 1
        j = i+1
        while j < len(remaining_days) and remaining_days[i] >= remaining_days[j]:
            cnt += 1
            j += 1
        answer.append(cnt)
        cnt=0
        i = j

    return answer

solution([95, 90, 99, 99, 80, 99], [1, 1, 1, 1, 1, 1])
```
