__3  __
- 3  , 3     
    - [0 => 1]
    - [1 => 2]
    - [2 => 4]
- but,     
-    , 4          

```python
n = 11

def solution(n):
    num = ['1','2','4']
    answer = ""


    while n > 0:
        
        n -= 1
        answer = num[n % 3] + answer
        n //= 3

    return answer

solution(n)
```
