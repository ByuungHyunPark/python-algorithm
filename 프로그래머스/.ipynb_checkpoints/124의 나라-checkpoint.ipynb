__3진수를 활용하여 풀이__
- 3진수까지는 생각을 하였지만, 3진수로 변환 후 값을 변환시켜서 풀이
    - [0 => 1]
    - [1 => 2]
    - [2 => 4]
- but, 생각보다 원하는데로 코드가 전혀 안됐음
- 전체적인 흐름은 이해를 하였으나, 4가 들어가야 하는 경우에 대해서는 이상한 결과가 나와서 구글링을 통해 해결

```python
n = 11

def solution(n):
    num = ['1','2','4']
    answer = ""


    while n > 0:
        
        n -= 1
        answer = num[n % 3] + answer
        n //= 3

    return answer

solution(n)
```
