##### 나의 풀이

```python
priorities = [1, 1, 9, 1, 1, 1]
location = 0
# return = 5

def solution(priorities, location):
    
    import collections
    
    answer = 0
    
    #deque가 속도측면에서 엄청 효율적이라고 함
    deq = collections.deque([(x, y) for x, y in enumerate(priorities)])
    
    
    while True:
        
        # deque 맨 왼쪽값이 최대값인지 확인
        if deq[0][1] == max(deq, key = lambda x : x[1])[1]:
            answer += 1
            
            if deq.popleft()[0] == location:
                return answer
            
        else:
            deq.append(deq.popleft())
            
    return answer

solution(priorities, location)
```

##### deque를 통해 index가 존재하는 list 형태로 문제를 풀이
- 문서의 대기목록이 있을때 맨 앞부터 탐색 => deque 맨 앞 인자부터 max인지 확인
- 현재 deque의 max값이면, pop 후, 그 값이 찾고자 하는 location이면 그 순위를 return
- 현재 deque의 max값이지만 location이 아니면, 최대값을 pop
- max가 아니라면, 뒤로 넘겨줌

위같은 순서를 계속 반복해가며, location에 위치한 값이 최대값이 되는 순서를 return 해주고자 함

```python
priorities = [2,1,3,2]
location = 2  
# return = 1

def solution(priorities, location):
    answer = 0
    from collections import deque
    
    deq = deque([(i, p) for i, p in enumerate(priorities)])
    
    while True:
        print('#inter{} "s deq : '.format(answer), deq)
        item = deq.popleft()
        
        if deq and max(deq)[0] > item[0]:
            deq.append(item)
        else:
            answer += 1
            if item[1] == location:
                break
                
    return answer
         
solution(priorities, location)
```

__참고 풀이__
- 전체적인 맥락은 유사하다. 하지만, deque에서 max값에 key 설정이 중요하단것을 알게됨

```python
def solution(priorities, location):
    answer = 0
    check = [ (x, y) for x, y in enumerate(priorities)]
    index = 0
    
    while True:
        if check[index][1] == max(check, key=lambda x:x[1])[1]:
            answer += 1
            if check.pop(index)[0] == location:
                return answer

        else:
            check.append(check.pop(index))

    return answer

solution(priorities, location)
```
