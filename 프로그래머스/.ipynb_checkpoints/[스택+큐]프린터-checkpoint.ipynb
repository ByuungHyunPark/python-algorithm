```python
priorities = [1, 1, 9, 1, 1, 1]
location = 0
# return = 5

def solution(priorities, location):
    
    import collections
    
    answer = 0
    
    #deque    
    deq = collections.deque([(x, y) for x, y in enumerate(priorities)])
    
    
    while True:
        
        # deque    
        if deq[0][1] == max(deq, key = lambda x : x[1])[1]:
            answer += 1
            
            if deq.popleft()[0] == location:
                return answer
            
        else:
            deq.append(deq.popleft())
            
    return answer

solution(priorities, location)
```

```python
priorities = [2,1,3,2]
location = 2  
# return = 1

def solution(priorities, location):
    answer = 0
    from collections import deque
    
    deq = deque([(i, p) for i, p in enumerate(priorities)])
    
    while True:
        print('#inter{} "s deq : '.format(answer), deq)
        item = deq.popleft()
        
        if deq and max(deq)[0] > item[0]:
            deq.append(item)
        else:
            answer += 1
            if item[1] == location:
                break
                
    return answer
         
solution(priorities, location)
```

```python
def solution(priorities, location):
    answer = 0
    check = [ (x, y) for x, y in enumerate(priorities)]
    index = 0
    
    while True:
        if check[index][1] == max(check, key=lambda x:x[1])[1]:
            answer += 1
            if check.pop(index)[0] == location:
                return answer

        else:
            check.append(check.pop(index))

    return answer

solution(priorities, location)
```
