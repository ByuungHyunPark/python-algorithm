```python
# input
n = int(input())
m = []
for _ in range(n):
    m.append(list(map(int, input().split())))
    

rank_list = []
    
# start
all_weight_list = [weight for weight in [wl[0] for wl in m]]
all_height_list = [height for height in [wl[1] for wl in m]]

for i in range(n):
    count = 1
    i_w = m[i][0]
    i_h = m[i][1]
    
    not_include_weight_list = all_weight_list[:i] + all_weight_list[i+1:]
    not_include_height_list = all_height_list[:i] + all_height_list[i+1:]
    
    for j,k in zip(not_include_weight_list, not_include_height_list):
        if (i_w < j) & (i_h < k):
            count += 1
    
    rank_list.append(count)
    
print(' '.join([str(i) for i in rank_list]))
```

```python
num_student = int(input())
student_list = []

for _ in range(num_student):
    weight, height = map(int, input().split())
    student_list.append((weight, height))

for i in student_list:
    rank = 1
    for j in student_list:
        if i[0] < j[0] and i[1] < j[1]:
                rank += 1
    print(rank, end = " ")
```
